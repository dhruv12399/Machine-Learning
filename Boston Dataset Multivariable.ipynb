```python
from sklearn import datasets,model_selection
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

boston = datasets.load_boston()
X = boston.data
Y = boston.target

df = pd.DataFrame(X)
df.columns = boston.feature_names
print(df.columns)
k=0
for i in  range(13):
    for j in range(i,13):
        first = df.columns[i]
        second = df.columns[j]
        name =  first + "_" + second
        df[name] = df[first]*df[second]
#         k+=1
# print(k)    
X2 = df.values
# X2.shape
print(df.describe())

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, random_state = 0)
X2_train, X2_test, Y2_train, Y2_test = model_selection.train_test_split(X2,Y, random_state = 0)

alg1 = LinearRegression()
alg2 = LinearRegression()

alg1.fit(X,Y)
alg2.fit(X2,Y)
print(alg1.coef_)
Y_pred = alg1.predict(X_test)
Y2_pred = alg2.predict(X2_test)

train_score = alg1.score(X_train,Y_train)
test_score = alg1.score(X_test,Y_test)
train_score2 = alg2.score(X2_train,Y2_train)
test_score2 = alg2.score(X2_test,Y2_test)

print("Training Score 1: ", train_score)
print("Testing Score 1: ", test_score)

print("Training Score 2: ", train_score2)
print("Testing Score 2: ", test_score2)
```
