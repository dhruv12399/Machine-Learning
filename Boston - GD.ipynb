```python
import numpy as np
import pandas as pd
# from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

def step_gradient(data,lr,m):
    N = len(data)
    n = len(m)
    m_slope = np.zeros(n)
    m_new = np.zeros(n)
    for i in range(N):
        a = np.dot(m,data[i][:-1])
        for j in range(n):
            x = data[i,j]
            y = data[i,n]
            m_slope[j] += (-2/N)*(y-a)*x
            m_new[j] = m[j]-lr*m_slope[j]
    return m_new


def cost(data,m):
    total_cost = 0
    N = len(data)
    n = len(m)
    for i in range(N):
        a = np.dot(m,data[i][:-1])
        y = data[i,n]
        total_cost += (1/N)*((y-a)**2)
    return total_cost

def gd(data,lr,ni):
    m=np.zeros(len(data[0][:-1]))
    for i in range(ni):
        m = step_gradient(data,lr,m)
        print("Cost ",i, " ", cost(data,m))
    return m

def predict(m,test_data):
    c = m[-1]
    m = m[:-1]
    Y_pred = []
    for i in range(len(test_data)):
        Y_pred.append(np.dot(test_data[i],m))
        Y_pred[i] += c
    return Y_pred

def run():
    boston = np.loadtxt("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/Boston/0000000000002417_training_boston_x_y_train.csv", delimiter = ",")
    boston_test = np.loadtxt("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/Boston/0000000000002417_test_boston_x_test.csv", delimiter = ",")
    print("boston ",boston.shape)
    X = boston
    input = X[:,:-1]
    Y = X[:,-1]
#     X = np.insert(boston,-1,1,axis=1)
#     boston_test = np.insert(boston_test,-1,1,axis=1)
    print("boston_test ",boston_test.shape)
#     print("X (after inserting 1) ",X.shape)
    # print(X)
    df = pd.DataFrame(input)
    df.columns  = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT']

    for i in  range(13):
        for j in range(i,13):
            first = df.columns[i]
            second = df.columns[j]
            name =  first + "_" + second
            df[name] = df[first]*df[second]

#     scalar = preprocessing.StandardScaler()
    # print(X[:,:-1].shape)
    
    input = df.values
#     X = input[:,:-1]
#     Y = input[:,-1]
    print("input ",input.shape)
    print("Y " ,Y.shape)
#     scalar.fit(input)
    print("input after taking degree 2 parameters :",input.shape)
#     input = np.insert(input,-1,1,axis=1)
    print("Y " ,Y.shape)
#     scalar.fit(input)
    lr = 0.013
    ni = 100
#     input = scalar.transform(input)
#     print("transformed input :", input.shape)
    
    Y = Y.tolist()
    input = input.tolist()
    for i in range(len(input)):
        input[i].append(Y[i])
    input = np.array(input)
    Y = np.array(Y)
    input = np.insert(input,-1,1,axis=1)
    print("final input :",input.shape)
    
#     print("input afet appending Y & 1 :",input.shape)
    m = gd(input,lr,ni)
    print("m :",m.shape)
    print(m)
#     boston_test = scalar.transform(boston_test)
    print("boston_test :",boston_test.shape)
    df2 = pd.DataFrame(boston_test)
    df2.columns  = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT']

    for i in  range(13):
        for j in range(i,13):
            first = df2.columns[i]
            second = df2.columns[j]
            name =  first + "_" + second
            df2[name] = df2[first]*df2[second]
            
    test_input = df2.values
#     test_input = scalar.transform(test_input)
    print("test_input :",test_input.shape)
#     boston_test = np.insert(boston_test,-1,1,axis=1)
#     print("boston_test after appending 1 :", boston_test.shape)
    Y_final = predict(m,test_input)
    print(len(Y_final))
    print(Y_final)
#     np.savetxt("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/Boston/predictions3.csv",Y_final)

run()
```
