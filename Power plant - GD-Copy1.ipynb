```python
import numpy as np
import pandas as pd
# from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

def step_gradient(data,lr,m):
    N = len(data)
    n = len(m)
    C = 500
    m_slope = np.zeros(n)
    m_new = np.zeros(n)
    for i in range(N):
        a = np.dot(m,data[i][:-1])
        for j in range(n):
            x = data[i,j]
            y = data[i,n]
            m_slope[j] += (-2*C/N)*(y-a)*x
            m_slope[j] += (1/N)*m[j]
            m_new[j] = m[j]-lr*m_slope[j]
    return m_new


def cost(data,m):
    total_cost = 0
    N = len(data)
    n = len(m)
    C = 500
    for i in range(N):
        a = np.dot(m,data[i][:-1])
        y = data[i,n]
        total_cost += (C/N)*((y-a)**2)
        total_cost += (0.5/N)*(m**2).sum()
    return total_cost

def gd(data,lr,ni):
    m=np.zeros(len(data[0][:-1]))
    for i in range(ni):
        m = step_gradient(data,lr,m)
        print("Cost ",i, " ", cost(data,m))
    return m

def predict(m,test_data):
    c = m[-1]
    m = m[:-1]
    Y_pred = []
    for i in range(len(test_data)):
        Y_pred.append(np.dot(test_data[i],m))
        Y_pred[i] += c
    return Y_pred

def run():
    boston = np.loadtxt("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/New Folder/0000000000002419_training_ccpp_x_y_train.csv", delimiter = ",")
    boston_test = np.loadtxt("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/New Folder/0000000000002419_test_ccpp_x_test.csv", delimiter = ",")
    print("boston ",boston.shape)
    X = boston
    poly = preprocessing.PolynomialFeatures(7)

#     X = np.insert(boston,-1,1,axis=1)
#     boston_test = np.insert(boston_test,-1,1,axis=1)
    print("boston_test ",boston_test.shape)
#     print("X (after inserting 1) ",X.shape)
    # print(X)
    input = X[:,:-1]
    Y = X[:,-1]
    scalar = preprocessing.StandardScaler()
    scalar.fit(input)
    input = scalar.transform(input)
    
#     df = pd.DataFrame(input)
#     df.columns  = ['TAX','PTRATIO', 'B', 'LSTAT']
    
#     for i in  range(4):
#         for j in range(i,4):
#             first = df.columns[i]
#             second = df.columns[j]
#             name =  first + "_" + second
#             df[name] = df[first]*df[second]
    
#     input = df.values
    input = poly.fit_transform(input)

    # print(X[:,:-1].shape)
#     input = X[:,:-1]
#     Y = X[:,-1]
    print("input ",input.shape)
    print("Y " ,Y.shape)
    
    lr = 0.000004
    ni = 10
    
    print("transformed input :", input.shape)
    
    Y = Y.tolist()
    input = input.tolist()
    for i in range(len(input)):
        input[i].append(Y[i])
    input = np.array(input)
    Y = np.array(Y)
    
    input = np.insert(input,-1,1,axis=1)
    print("input afet appending Y & 1 :",input.shape)
    
    m = gd(input,lr,ni)
    print(m.shape)
    
    boston_test = scalar.transform(boston_test)
    boston_test = poly.transform(boston_test)

#     df2 = pd.DataFrame(boston_test)
#     df2.columns  = ['TAX','PTRATIO', 'B', 'LSTAT']

#     for i in  range(4):
#         for j in range(i,4):
#             first = df2.columns[i]
#             second = df2.columns[j]
#             name =  first + "_" + second
#             df2[name] = df2[first]*df2[second]
            
#     boston_test = df2.values
    
    
    print(boston_test.shape)
#     boston_test = np.insert(boston_test,-1,1,axis=1)
#     print("boston_test after appending 1 :", boston_test.shape)
    Y_final = predict(m,boston_test)
    print(len(Y_final))
    print(Y_final)
    np.savetxt("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/New Folder/predictions2.csv",Y_final)

run()
```
